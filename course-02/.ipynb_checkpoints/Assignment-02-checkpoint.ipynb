### Get the frequences of words

````python
from collections import Counter
````

````python
words_count = Counter(valida_tokens)
````

````python
frequences_all = [f for w,f in words_count.most_common()]
````

````python
frequences_sum = sum(frequences_all)
````

````python
frequences_sum
````

````python
def get_prob(word):
    esp = 1/ frequences_sum
    if word in words_count:
        return words_count[word] / frequences_sum
    else:
        return esp
````

````python
get_prob('小米')
````

````python
def product(pi):
    return reduce(lambda x1,x2 : x1*x2, pi)
````

````python
product([1,2,3,4,5,6])
````

````python
def language_model_one_gram(string):
    words = cut(string)
    return product([get_prob(w) for w in words])
````

````python
language_model_one_gram('深圳是离香港最最近的城市')
````

````python
language_model_one_gram('明天我要开奔驰去拉板砖')
````

## Get the combination probability of 2-gram

$$ P_r( w_2 | w_1) = \frac{P_r(w_1 \cdot w_2)}{P_r(w_1} $$

````python
valid_tokens = [str(t) for t in valida_tokens]
all_2_gram_words = [''.join(valid_tokens[i:i+2]) for i in range(len(valid_tokens[:-2]))]
````

````python
all_2_gram_words[:10]
````

````python
_2_gram_sum = len(all_2_gram_words)
_2_gram_counter = Counter(all_2_gram_words)
````

````python
_2_gram_counter.most_common(4)
````

### 计算 $ P_r(w_1 \cdot w_2)$

````python
def get_combination_prob(w1,w2):
    if w1 + w2 in _2_gram_counter: return _2_gram_counter[w1+w2] / _2_gram_sum
    else:
        return 1/ _2_gram_sum
````

````python
get_combination_prob('去', '北京')
````

````python
get_combination_prob('小米','飞机')
````

````python
get_combination_prob('小米','手机')
````

### 计算 $ P_r( w_2 | w_1) = \frac{P_r(w_1 \cdot w_2)}{P_r(w_1} $

````python
def get_prob_2_gram(w1,w2):
    return get_combination_prob(w1,w2) / get_prob(w1)
````

````python
get_prob_2_gram('去','深圳')
````

````python
get_prob_2_gram('去','北京')
````

````python
def language_model_of_2_gram(sentence):
    sentence_probability = 1
    words = cut(sentence)
    
    for i,word in enumerate(words):
        if i == 0:
            prob = get_prob(word)
        else:
            previous = words[i-1]
            prob = get_prob_2_gram(previous,word)
        sentence_probability *= prob
    return sentence_probability
````

#### 9. What't the 2-gram models;
Ans: when we predict the probability of current words, we should consider the previous word. That can be represented using the following equation. combination probability:
$$ P_r( w_2 | w_1) = \frac{P_r(w_1 \cdot w_2)}{P_r(w_1} $$

the probability of the sentence: $ P(sentence) = P(w_1,w_2,\cdots ,w_n)$, we use the following equation to substitue the probability of whole sentence, this is the 2-gram language model.

$$ P(w_1) \cdot \prod_{i=2}^{n} P(w_i|w_{i-1}) $$
