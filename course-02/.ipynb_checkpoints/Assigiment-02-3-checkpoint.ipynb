### Get the frequences of words

```python
from collections import Counter
```

```python
all_frequences = Counter(valida_tokens)
```

```python
all_frequences.most_common(10)
```

```python
frequence_all_words = [f for w,f in all_frequences.most_common()]
```

```python
frequence_sum = sum(frequence_all_words)
```

```python
def get_prob_one_word(word):
    esp = 1/frequence_sum
    if word in all_frequences:
        return all_frequences[word] / frequence_sum
    else:
        return esp
```

```python
from functools import reduce
def product(numbers):
    return reduce(lambda x1,x2: x1 * x2,numbers)
```

```python
def language_model_one_gram(string):
    words = cut(string)
    return product([get_prob_one_word(w) for w in words])
```

```python
language_model_one_gram('广交会下个月举办')
```

```python
language_model_one_gram('长征火箭下周发射')
```

### compute the combination probability
$$ P_r( w_2 | w_1) = \frac{P_r(w_1 \cdot w_2)}{P_r(w_1)} $$

#### 处理两个相邻的词

```python
valida_tokens[:10]
# [str(t) for t in valida_tokens[:12]]
```

```python
all_2_gram_words = [''.join(valida_tokens[i:i+2]) for i in range(len(valida_tokens[:-2]))]
```

```python
all_2_gram_words[:10]
```

#### compute the combination probability 
$Pr(w_1 \cdot w_2)$

```python
_all_2_gram_sum = len(all_2_gram_words)
_all_2_gram_counter = Counter(all_2_gram_words)

def get_prob_combination(w1,w2):
    esp = 1/ _all_2_gram_sum
    if w1+w2 in _all_2_gram_counter: return _all_2_gram_counter[w1+w2] /_all_2_gram_sum
    else:
        return esp
```

```python
get_prob_comb_2_word('去','北京')
```

```python
get_prob_comb_2_word('坐','火箭')
```

#### compute
$ P_r( w_2 | w_1) = \frac{P_r(w_1 \cdot w_2)}{P_r(w_1)} $

```python
def get_prob_2_gram(w1,w2):
    previous = w1
    return get_prob_combination(previous,w2) /get_prob_one_word(previous)
```

```python
def language_model_2_gram(string):
    sentence_pro = 1
    words = cut(string)
    for i,w in enumerate(words):
        if i == 0:
            prob = get_prob_one_word(w)
        else:
            previous = words[i-1]
            prob = get_prob_2_gram(previous,w)
        sentence_pro *= prob
    return sentence_pro
```

```python
language_model_2_gram('小明今天抽奖抽到了一个苹果手机')
```

```python
language_model_2_gram('小明今天开着波音飞机去拉板砖')
```

```python
language_model_2_gram('小明今天抽奖')
```

## Step 5: If we need to solve following problems, how can language model help us?

### Voice Recognization.

### Sogou pinyin input.

### Auto correction in search engine.
- 基于拼音的中文拼写纠错-同音字纠错
 然后对于同一个音的所有候选字计算出修正概率： 
$$ P（w|x）=P(x|w)\cdot P(w) $$
其中每个音匹配的不同字的状态转移概率为P（w）（基于统计获得），而P(x|w)则是对于不同字对应的正确字的转移概率（基于统计获得），最终可以得出最终的纠正后的结果。
- 不同音字的错误就只能依靠n-gram语言模型了

### Abnormal Detection.
用深度学习来训练，提取异常的特征，进行异常检测

### Compared to the previous learned parsing and pattern match problems. What's the advantage and disavantage of Probability Based Methods?

Advantage：
- 效率高。当数据量特别大的时候，基于模式匹配的方法在进行一一比较的时候，速度非常慢，很影响效率，基于统计概率的方法，在这点上速度要快很多，在模型训练好以后，不需要再处理整个文本的概率了，只需要计算当前句子的概率。
- 不用处理复杂的文本语料，程序开发周期短

Disadvantage：
 - 分词的准确率不高，
 - 不能识别新词
 - 语言歧义问题：（如：南京市长江大桥欢迎您，南京市市长叫江大桥？？）
