6. precision， recall，f1, auc 分别是什么意思？ 假设一个城市有 10000 人，有 30 个犯罪分子，警察抓到了 35 个人，其中 20 个是犯罪分子，请问这个警察的 precision, recall, f1,auc 分别是什么？ 

Ans:
+ Precision：表示预测为正样本中有多少个是真正的正样本，$20/35 = \frac{4}{7}$
+ Recall： 表示预测结果中，覆盖了多少正确的正样本， $20/30 = \frac{2}{3}$
+ F1-score: 是将precision 和 recall合并成一个指标，$F1_{score} = \frac{2*precision*recall}{precision+recall} = \frac{8}{13}$
+ AUC : 是ROC曲线对横轴的面积，ROC曲线：横轴是$FPR = \frac{FP}{TN+FP}$，纵轴是$TPR=\frac{TP}{TP+FN} =Recall$

8. 什么是 Overfitting， 什么是 Underfitting?

Ans:
+ Overfitting：过拟合，模型过度拟合了训练数据的内在联系，包括一些噪声数据，这样模型在训练数据上表现很好，在非训练数据上表现不好
+ Underfitting：欠拟合，模型没有学到数据的内在联系，由于模型太简单了，无法刻画一下复杂的情形。

12. Bias 和 Variance 有什么关系？ 他们之间为什么是一种 tradeoff 的？ 

Ans:
+ bias：偏差，描述的是通过学习拟合出来的结果的期望，与真实规律之间的差距
+ Variance：方差，描述的是通过学习拟合出来的结果自身的不稳定性
+ 联系：有监督学习中，预测误差的主要来源就是bias和variance，模型的性能取决于bias与variance的tradeoff
+ why？低的偏差，低的方差，才能使得模型更加准确，如何协调low bias 和low variance，就是要找到一个最佳平衡点，使得模型更加准确，如果模型复杂度大于平衡点，则模型的variance会偏高，模型倾向于过拟合，如果模型的复杂度小于平衡点，则模型bias会偏高，模型倾向于过拟合

13. Train， Validation，Test 数据集之间是什么关系？ 为什么要这么划分？ 

Ans:
+ train data：目的是训练模型的参数
+ validation：用来做模型选择，即做模型的最终优化及确定的
+ Test data：测试已经训练好的模型的推广能力
+ 训练数据和验证数据是相互独立的，使用的是不同的数据
+ 模型需要训练数据来拟合模型，训练模型的参数，而验证数据是用来调整模型参数从而选择最优模型，然后使用测试数据去评估模型的表现，但是并不会调整优化模型

16. Linear Regression中，Loss 函数怎么定义的？ 为什么要写成这样？ 什么是凸函数？ 优化中有什么意义？ 

Ans:
+ loss：平方损失函数$loss = \sum_{i=1}^n(y_i-\bar{y})^2$
+ 为什么使用平方：使用的是最小二乘的思想，利用平方来度量观测点与估计点的距离（远近），最小指的是参数值要保证各个观测点与估计点的距离的平方和达到最小
+ 凸函数：定义在某个向量空间的凸子集上的实值函数，其在函数图像上方的点集为一个凸集
+ 凸优化：**针对凸优化问题，能够比较有效的找到全局最优解**，因此现实生活中很多计算量接近无穷大的问题时，想办法将模型转换成凸优化问题，然而很多问题都不是凸优化问题，因此凸优化很重要。

+ 用直观的例子解释。比如我们在一座大山上的某处位置，由于我们不知道怎么下山，于是决定走一步算一步，也就是在每走到一个位置的时候，求解当前位置的梯度，沿着梯度的负方向，也就是当前最陡峭的位置向下走一步，然后继续求解当前位置梯度，向这一步所在文职沿着最陡峭最易下山的位置走一步，这样一步步走下去，一直走到觉得我们已经到了山脚。有可能是到了一个局部的山脚。

18. 一般在机器学习数量时，会做一个预处理（Normalization）， 简述 Normalization 的过程，以及数据经过 Normalization之后的平均值和标准差的情况。

Ans:
+ 是将数据按比例缩放，使之落到一个小的特定区间，
+ 求出数据的平均值$\bar{X}$和标准差$Std$，然后对每一个数据使用$X_{Nor} = \frac{x - \bar{X}}{Std} $进行标准化。
+ 标准化后，数据被映射到某种特定的分布中，均值为0，方差为1

19. Logstic Regression 的 Logstic 是什么曲线，被用在什么地方？ 

Ans:
+ 逻辑回归是一个S型的曲线，能够把值映射到（0，1）之间，$y = \frac{1}{1+e^{-x}}$
+ logstic regression常见的应用场景就是预测概率，如根据一个人的一些特征预测其患心脏病的概率等

20. Logstic Regression 的 Loss 函数 Cross Entropy 是怎么样的形式？ 有什么意义？ 

Ans
+ Cross Entropy:  $loss = - \sum(y_i log\hat{y})$
+ 描述了两个不通过概率分布$y_i $和$ \hat{y}$的差异程度，两个分布差异越大，则交叉熵的差异越大

```python
def token(string):
    return re.findall(r'[\d|\w]+', string)

def cut(string):
    return ' '.join(jieba.cut(string))

def Preprocess_token(content):
    news_content = [token(n) for n in content]
    news_content = [' '.join(x) for x in news_content]
    news_content = [cut(y) for y in news_content]
    return news_content
```
