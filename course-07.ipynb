1. Summarize the reasons of overfitting and underfitting. Put them in github repository.



The Reason of overfitting:
1. the model is too complex;
2. the train data has much noise and model learns the noise in the training data to the extent that it negatively impacts the performance of the model on new data.

The reason of underfitting:
1. the model is too simple and it can neither model the training data and generalize to new data.
